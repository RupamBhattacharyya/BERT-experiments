```python
df_train = pd.read_csv("/content/drive/My Drive/IIIT Creative Content/Diaalar Lab/TokenizerInfluence/train.csv")
#df = df_train.iloc[:, :-1]
print(df_train.head())

from sklearn import preprocessing
labeler = preprocessing.LabelEncoder()
df2 = pd.DataFrame(df_train["Questions"])
df2 = df2.applymap(str)
df2['label'] = labeler.fit_transform(df_train["label"])
print(df2.head())
```

```python
model = ClassificationModel('bert', 'bert-base-multilingual-uncased', num_labels=6, use_cuda=True, args={
    'reprocess_input_data': True,
    'use_cached_eval_features': False,
    'overwrite_output_dir': True,
    'num_train_epochs': 10
})
model.train_model(df2)
```

```python
orig_test = pd.read_csv("/content/drive/My Drive/IIIT Creative Content/Diaalar Lab/TokenizerInfluence/val.csv")
eval_frame = pd.DataFrame(orig_test["Questions"])
eval_frame = eval_frame.applymap(str)
eval_frame['label'] = labeler.fit_transform(orig_test["label"])
print(eval_frame.head())
```

```python
print(len(eval_frame))
```

```python
result, model_outputs, wrong_predictions = model.eval_model(eval_frame)
bads = {}
for pred in wrong_predictions:
    if pred.label in bads:
        bads[pred.label] += 1
    else:
        bads[pred.label] = 1
print("wrong predictions (by input I think?):")
print(bads)
```
